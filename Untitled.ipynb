```python
from sklearn.preprocessing import OneHotEncoder
cat_cols = df.select_dtypes(include='object').columns
oe = OneHotEncoder(drop='first', sparse_output=False) 

cat_encoded = oe.fit_transform(df[cat_cols])
cat_encoded_df = pd.DataFrame(cat_encoded, columns=oe.get_feature_names_out(cat_cols))

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df_final = pd.concat([df[num_cols], cat_encoded_df], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_final.iloc[:,:-1]
y=df_final['Churn_Yes']
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
lo=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
ga=GaussianNB()
l=[lo,knn,dt,ga]
```

```python
for model in l:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    print("==============================================")
    print(f"Model: {model.__class__.__name__}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)*100}")
    report = classification_report(y_test, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    print(report_df)
    print("Confusion_Matrix")
    print("-----------------------------------------------")
    print(confusion_matrix(y_test, y_pred))
    print("=================================================")
```

```python
print("Train_Score",lo.score(x_train,y_train)*100)
print("Test_Score",lo.score(x_test,y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rdf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='sqrt',
    random_state=42
)
rdf.fit(x_train, y_train)
pre=rdf.predict(x_test)
print("Accuracy_score : ",accuracy_score(y_test,pre)*100)
print("Train_Score : ",rdf.score(x_train,y_train)*100)
print("Test_Score : ",rdf.score(x_test,y_test)*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(estimator=rdf,n_estimators=100,learning_rate=0.05,random_state=42)

ada.fit(x_train, y_train)

pred = ada.predict(x_test)

print("Accuracy_score :", accuracy_score(y_test, pred)*100)
print("Train_Score :", ada.score(x_train, y_train)*100)
print("Test_Score :", ada.score(x_test, y_test)*100)
```
